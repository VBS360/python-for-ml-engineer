## ❓ 1. Explain your project end-to-end

👉 (100% will be asked)

Sure.

I built a health report analyzer with a chatbot interface to help users understand complex medical reports in simple language and ask follow-up questions.

The system works in three main stages:

**1. Document Processing Pipeline:**
- When a user uploads a report (PDF/Word/Text),
- We use Tesseract OCR to extract text from the document
- The extracted text is then chunked and converted into embeddings using OpenAI
- These embeddings are stored in a vector database

**2. Summary Generation:**
- We retrieve relevant chunks from the vector database
- Pass them to the LLM
- The model generates a simplified health summary and recommendations

**3. Chatbot (RAG-based Q&A):**
- When a user asks a question:
    - The query is converted into an embedding
    - We perform similarity search in the vector DB
    - Retrieve the most relevant chunks
    - Pass those chunks + user query to the LLM
- The LLM generates a context-aware response based on the report

 **Why this approach**
- Helps reduce hallucination using RAG
- Makes medical data more interpretable for users
- Enables interactive exploration instead of static reports

## ❓ 2. How does your chatbot work internally?

- Internally, my chatbot follows a RAG-based pipeline:

1. User Query Processing:
- User asks a question
- The query is converted into an embedding using an embedding model

2. Similarity Search
- We perform top-k similarity search in the vector database
- Retrieve the most relevant text chunks related to the query

3. Prompt Construction
- We combine:
-- User query
-- Retrieved context (chunks)

- Into a structured prompt like:

“Answer based only on the provided context”

4. Response Generation:
- This prompt is sent to the LLM
- The model generates a context-aware response grounded in the retrieved data

5. Fallback Handling
- If no relevant data is found,
- The system responds with:

“I don’t have enough information from your report”

- So overall, the chatbot ensures responses are accurate, contextual, and grounded in user data.

## ❓ Why did you choose RAG instead of fine-tuning a model on medical data?

- RAG was a better choice than fine-tuning for three main reasons:

1. Personalization without retraining:
- Each user uploads a different health report
- With RAG, we can dynamically retrieve user-specific data
- Fine-tuning would not scale for individual users

2. Cost and scalability:
- Fine-tuning a model for every user would be computationally expensive and impractical
- RAG allows us to use a single pre-trained model efficiently across all users

3. Real-time and updatable knowledge:
- In RAG, we can simply update the vector database with new reports
- No need to retrain the model every time new data comes in

4. Reduced hallucination:
- Since responses are grounded in retrieved documents,
- The model gives more accurate and context-specific answers

- That’s why RAG was more suitable for this use case.

- If the system retrieves incorrect or irrelevant chunks, the issue is usually in the retrieval stage, not the LLM.

- I would handle it in multiple ways:

1. Improve similarity search quality:
- Use better embedding models

- Ensure proper chunking of documents (not too large or too small)

2. Add re-ranking:
- After retrieving top results, use a re-ranking model to select the most relevant chunks

3. Query refinement:
- Improve how user queries are processed
- Sometimes rewrite or expand queries for better matching

4. Handle out-of-scope queries:

- Add system prompts to ensure:

“Answer only from provided context. If not found, say you don’t know.”

5. Reduce hallucination:
- Strict prompting + context grounding
- Avoid answering beyond retrieved data

- So overall, I would focus on improving retrieval quality and prompt control, rather than fine-tuning the base model.

## ❓ How do you decide chunk size when storing documents in a vector database?

- Chunk size selection is important because it affects both retrieval quality and context relevance.

1. If chunks are too large:
- They may contain irrelevant information
- This reduces precision in similarity search
- The model may get confused with too much context

2. If chunks are too small:
- Important context may be lost
- The retrieved chunk may not have enough information to answer the query

3. How I would decide:
- Start with a moderate chunk size (e.g., 300–500 tokens)
- Use overlapping chunks to preserve context
- Experiment and evaluate based on:
- Retrieval accuracy
- Response quality

4. In my project:
- Since health reports have structured data,
- I would chunk based on logical sections (like test results, diagnosis, etc.) instead of random splits

- So overall, chunking is a balance between context completeness and retrieval precision.

## ❓ 4. Why did you use embeddings?

- Embeddings are used because they capture the semantic meaning of text, not just exact keyword matches.

1. Semantic Understanding:
- Text-based search relies on exact keyword matching
- But embeddings represent meaning, so:
“high blood sugar”
“diabetes”
👉 can be understood as similar

2. Better Similarity Search:
- In vector databases, we compare embeddings using distance metrics
- This allows us to find conceptually similar content, even if wording is different

3. Efficiency in Retrieval:
- Numerical vectors enable fast and scalable similarity search
- Which is essential for real-time systems

4. Why not plain text search:
- Keyword search would miss relevant results if wording differs
- Embeddings solve this by enabling context-aware retrieval

- So embeddings help us move from keyword matching → meaning-based search.

## ❓ Which similarity metric is used in vector search? Cosine, Euclidean… what and why?
- Yes, we typically use cosine similarity for vector search.

- **Why cosine similarity?**
- It measures the angle between two vectors, not their magnitude
- This helps us compare the direction (i.e., meaning) of the text
- Why is that important?

- In embeddings, the direction represents semantic meaning
- Two texts with similar meaning will have vectors pointing in similar directions, even if their magnitude differs

- **Why not Euclidean distance?**
- Euclidean considers magnitude as well
- But magnitude is less important than meaning in NLP tasks
- So cosine similarity gives better semantic matching

- That’s why cosine similarity is preferred in most vector search systems.

- Too small → loss of context
- Too large → noisy retrieval
- Optimize chunk size + add overlap

d. Retrieval tuning
- Adjust top-k values
- Add re-ranking to improve relevance

So overall, I would ensure a safe fallback first, and then improve the retrieval pipeline.

## ❓ 11. What is an LLM?
- An LLM, or Large Language Model, is a deep learning model—typically based on the Transformer architecture—that is trained on massive amounts of text data.

- *How it works*
- It learns to predict the next word (token) in a sequence
- By doing this repeatedly during training, it captures:
-- Language patterns
-- Context
-- Relationships between words

- What it can do
-- Generate human-like text
-- Answer questions
-- Summarize content
-- Assist in conversations

- Why it’s powerful
-- Because it understands context and semantics, not just keywords
-- And can generalize across many tasks without task-specific training

- So in simple terms, an LLM is a model that learns language patterns at scale and can generate meaningful text based on context.
---

## ❓ What is a token?
- A token is the smallest unit of text that an LLM processes.

- **What does that mean?**
- A token can be:
-- A word
- Part of a word (subword)
- Or even punctuation
- Example
“ChatGPT is great” → might be split into:
“Chat”, “GPT”, “ is”, “ great”

- Why tokens matter
- LLMs process input in terms of tokens, not full sentences

- They determine:
- Cost (API pricing)
- Context window limits
- Model performance

- So tokens are the basic building blocks of how LLMs read and process text.

## ❓ 14. What is temperature in LLM?
- Temperature is a parameter in LLMs that controls the randomness or creativity of the generated output.

- How it works
-- The model assigns probabilities to possible next words
-- Temperature adjusts how confident or random the selection is

- *Low temperature (e.g., 0.2)*
-- More deterministic
-- Safer, more factual answers
-- Less variation

- *High temperature (e.g., 0.8–1.0)*
-- More creative and diverse
-- Higher chance of variation
-- But can increase errors or hallucination

- In my project
-- I would use low temperature
-- Because health-related responses should be accurate and reliable

- So temperature controls how predictable vs creative the model’s output is.
---

## ❓ 15. What is hallucination?
- Hallucination in LLMs refers to when a model generates incorrect or fabricated information that sounds confident but is not grounded in reality or provided data.

- *Why it happens*
- LLMs are trained to predict the next token, not verify facts
- They rely on patterns in data, not true understanding

- Example
- If asked a question without enough context,
- The model may still generate a plausible but incorrect answer

- How to reduce it
-- Use RAG to provide grounded context
-- Apply strict prompting
-- Set lower temperature

- So hallucination is essentially the model being confident but wrong.
---

## ❓ 16. What is context window?
- A context window is the maximum number of tokens a model can process at once, including both input and output.

- **What does it include?**
-- User query
-- System prompt
-- Retrieved context (from RAG, if used)
-- Previous conversation (in chat systems)

- **Why it matters**
- If input exceeds the context window,
→ older information gets truncated or ignored
- It limits how much information the model can consider at once

-- Example
- If a model has a 4K token context window,
- It can only process around 4000 tokens total (input + output)
- Relation with RAG
- RAG helps bring external data
- But that data must still fit inside the context window

- So context window is the model’s memory capacity, not the retrieval system itself.
