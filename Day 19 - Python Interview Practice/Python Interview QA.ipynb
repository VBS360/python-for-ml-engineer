#### Reverse a String

```python
def reverse_str(a):
    rev_a = ""
    for x in range (len(a)):
        rev_a += a[-x-1]
    return rev_a

a = "String"
reverse_str(a)
```

#### Check Palindrome

```python
def is_palindrome(a):
    rev_a = ""
    for x in range (len(a)):
        rev_a += a[-x-1]
    if rev_a == a:
        return True
    else:
        return False

a = "hello"
is_palindrome(a)
```
