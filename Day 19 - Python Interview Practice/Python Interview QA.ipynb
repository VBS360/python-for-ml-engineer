#### Q1: Reverse a String

```python
def reverse_string(s):
    rev_s = ""
    for x in range (len(s)):
        rev_s += s[-x-1]
    return rev_s

a = "String"
reverse_string(a)
```

#### Q2: Check Palindrome

```python
def is_palindrome(s):
    rev_s = ""
    for x in range (len(s)):
        rev_s += s[-x-1]
    if rev_s == s:
        return True
    else:
        return False

s = "hello"
is_palindrome(s)
```

#### Q4: Count Frequency of Elements

```python
def frequency_count(lst):
    freq = {}
    for x in lst:
        if x in freq:
            freq[x] += 1
        else:
            freq[x] = 1
    return freq

lst = [1,2,3,4,5,3,2,1]
frequency_count(lst)
```

#### Q5: Find Second Largest Element

```python
def second_largest(lst):
    largest = float('-inf')
    second = float('-inf')

    for num in lst:
        if num > largest:
            second = largest
            largest = num
        elif num > second and num != largest:
            second = num

    return second

lst = [10,20,4,45,99,15]
second_largest(lst)
```
