# 📅 Day 12 – XGBoost, Threshold Tuning & Interview Preparation

## 🚀 Overview

Day 12 focused on improving model performance using advanced algorithms and developing a deeper understanding of evaluation techniques. The key highlight was applying XGBoost and optimizing predictions using threshold tuning. Additionally, this day emphasized interview preparation through real-world ML questions.

---

## 🎯 Key Learnings

### 1. 🌳 XGBoost (Extreme Gradient Boosting)

* Learned how boosting works by building models sequentially.
* Understood how XGBoost improves performance by correcting previous errors.

#### 🔍 Key Takeaways:

* Performs well on structured/tabular data
* Includes regularization to reduce overfitting
* Often outperforms Random Forest

---

### 2. ⚖️ Model Comparison (Random Forest vs XGBoost)

| Metric   | Random Forest | XGBoost        |
| -------- | ------------- | -------------- |
| F1 Score | 0.34          | 0.15 (default) |
| ROC-AUC  | 0.68          | 0.74           |

#### 💡 Insight:

* Random Forest performed better initially in detecting defaulters (higher F1)
* XGBoost showed better overall classification ability (higher ROC-AUC)

---

### 3. 🎯 Threshold Tuning (Critical Learning ⭐)

* Default classification threshold = 0.5
* Lowering threshold improved recall and F1-score

#### 🔍 Experiment:

| Threshold | F1 Score |
| --------- | -------- |
| 0.1       | 0.32     |
| 0.2       | **0.35** |
| 0.3       | 0.30     |
| 0.4       | 0.25     |

#### 💡 Insight:

* Threshold = 0.2 gave best balance between precision and recall
* XGBoost became best performing model after tuning

---

### 4. 📊 ROC-AUC Interpretation

* Measures model’s ability to distinguish between classes
* Independent of threshold

#### 📌 Interpretation:

* 0.5 → Random
* 0.6–0.7 → Acceptable
* 0.7–0.8 → Good

---

### 5. 🧠 Business-Oriented Thinking

#### Key Insight:

* Identifying defaulters is more important than maximizing accuracy

#### Actionable Strategies:

* Use lower threshold to catch more defaulters
* Accept some false positives to reduce financial risk
* Apply risk-based lending policies

---

## 🔥 Interview Questions & Answers

---

### ❓ Q1: Explain your project

**Answer:**
Built a loan default prediction model using classification algorithms. Addressed class imbalance, applied multiple models, and improved performance using threshold tuning. XGBoost achieved best results with F1-score of 0.35 and ROC-AUC of 0.74.

---

### ❓ Q2: Why is accuracy not enough?

**Answer:**
The dataset is imbalanced, with ~90% non-defaulters. A model predicting all customers as non-defaulters would still achieve high accuracy but fail to detect actual defaulters. Metrics like F1-score and recall are more important.

---

### ❓ Q3: Why did XGBoost improve after threshold tuning?

**Answer:**
XGBoost outputs probabilities. Lowering the threshold increased recall by classifying more instances as defaulters, improving the balance between precision and recall, and thus increasing F1-score.

---

### ❓ Q4: How would you improve the model further?

**Answer:**

* Perform advanced hyperparameter tuning
* Handle imbalance using techniques like SMOTE
* Apply feature engineering
* Optimize threshold based on business needs
* Use cross-validation for robustness

---

### ❓ Q5: Is it a problem if a good customer is predicted as defaulter?

**Answer:**
Yes, it is a false positive and may lead to lost business. However, it is often acceptable to reduce false negatives (actual defaulters), which are more costly. The balance depends on business risk tolerance.

---
