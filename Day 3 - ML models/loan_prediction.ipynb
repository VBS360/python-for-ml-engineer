```python
import pandas as pd
```

##### Load dataset

```python
df = pd.read_csv("loan.csv")
df.head()
```

```python
df["Gender"] = df["Gender"].fillna(df["Gender"].mod)
df["Gender"] = df["Gender"].apply(lambda x: 0 if "Gender"=="Male" else 1)
df["Married"] = df["Married"].fillna(df["Married"].mod)
df["Married"] = df["Married"].apply(lambda x: 0 if "Married"=="No" else 1)
df["Dependents"] = df["Dependents"].fillna(df["Dependents"].mod)
df["Self_Employed"] = df["Self_Employed"].fillna(df["Self_Employed"].mod)
# df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].median, inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].median(), inplace=True)
df["Loan_Amount_Term"] = df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].mean)
df["Loan_Status"] = df["Loan_Status"].fillna(df["Loan_Status"].mod)
df["Credit_History"] = df["Credit_History"].fillna(df["Credit_History"].mod)
```
