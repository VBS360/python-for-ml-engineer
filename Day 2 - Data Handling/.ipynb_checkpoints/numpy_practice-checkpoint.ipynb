#### 🧱 Basics

```python
import numpy as np
```

#### Create 1D array

```python
arr1 = np.array([10,11,12])
print(arr1)
```

#### Create 2D array

```python
arr2 = np.array([[10,11,12],[4,5,6]])
print(arr2)
```

#### 📐 Shape & Dimensions

#### Print shape of arrays

```python
print(arr1.shape)
```

```python
print(arr1.shape)
```

#### Print number of dimensions (ndim)

```python
print(arr1.ndim)
```

```python
print(arr2.ndim)
```

#### 🔄 Reshaping

#### Convert 1D array → 2D

```python
reshaped = arr2.reshape(3,2)
print(reshaped)
```

#### 🔍 Indexing & Slicing

#### Get first row of 2D array

```python
print(arr2[0])
```

#### Get second column

```python
print(arr2[:,1])
```

#### Access a single element

```python
print(arr2[1,1])
```

```python
print(arr1[2])
```

#### ➕ Basic Operations

#### Find sum of array

```python
print(arr2.sum())       # sum of all elements
```

```python
print(arr2.sum(1))       # sum of raw elements
```

```python
print(arr2.sum(0))       # sum of column elements
```

#### Find mean

```python
print(arr2.mean())
```

```python
print(arr2.mean(0))
```

```python
print(arr2.mean(1))
```

#### Find max

```python
print(arr2.max())
```

```python
print(arr2.max(0))
```

```python
print(arr2.max(1))
```

#### ⚡ Element-wise Operations

#### Multiply array by 2

```python
arr2
```

```python
print(arr2*2)
```

```python
print(arr2[0]*2)
```

```python
print(arr2[1,1]*4)
```

#### Add 5 to all elements

```python
print(arr2+5)
```

#### 🎯 Filtering

#### Extract values greater than a number

```python
print(arr2[arr2>5])
```
