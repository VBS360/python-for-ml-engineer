# 🎯 SECTION 1: Prompt Engineering (Core – HIGH PRIORITY)

# 🧠 What is a Prompt Structure (in simple terms)

Standard Prompt Structure

Think of it as 5 building blocks:

### 🔹 1. Role (Who the AI is)

Sets behavior

👉 Example:

* “You are a data analyst”
* “You are an expert recruiter”

---

### 🔹 2. Task (What to do)

Core instruction

👉 Example:

* “Summarize the following text”
* “Classify this review as positive or negative”

---

### 🔹 3. Context (Background info)

Gives clarity

👉 Example:

* Target audience
* Domain info
* Extra details

---

### 🔹 4. Constraints (Rules)

Controls output

👉 Example:

* Word limit
* Tone (formal, simple)
* Do/Don’t instructions

---

### 🔹 5. Output Format (VERY IMPORTANT)

Ensures usable output

👉 Example:

* Bullet points
* JSON
* Table
* Step-by-step

# 🎯 1. What is prompt engineering?

**Answer:**

Prompt engineering is the process of designing and refining inputs given to an AI model to produce accurate, relevant, and structured outputs.
It involves defining clear instructions, adding context, setting constraints, and iteratively improving prompts based on the model’s responses.

👉 *(Optional add-on for impact)*
“I approach it as an iterative process where prompts are tested and optimized like test cases.”

---

# 🎯 2. What makes a good prompt?

**Answer:**

A good prompt is:

* **Clear and specific** (no ambiguity)
* **Well-structured** (role + task + constraints)
* **Context-aware** (provides necessary background)
* **Output-driven** (defines expected format)

👉
“In my experience, adding constraints and output format improves consistency the most.”

---

# 🎯 3. Difference between zero-shot, one-shot, and few-shot prompting?

**Answer:**

* **Zero-shot:** No examples, only instruction
* **One-shot:** One example is provided
* **Few-shot:** Multiple examples are provided to guide the model

👉
“Few-shot prompting is useful when the task requires pattern understanding or consistency.”

---

# 🎯 4. What is role-based prompting?

**Answer:**

Role-based prompting assigns a specific identity or role to the model to influence its behavior and response style.

👉 Example:
“You are a financial advisor…”

👉
“This helps improve relevance, tone, and domain-specific accuracy.”

---

# 🎯 5. What is chain-of-thought prompting?

**Answer:**

Chain-of-thought prompting encourages the model to explain its reasoning step-by-step before giving the final answer.

👉 Example:
“Explain your reasoning step-by-step…”

👉
“It improves performance in complex reasoning tasks like math, logic, or multi-step problems.”

---

# 🎯 6. How do you handle ambiguous prompts?

**Answer:**

I handle ambiguity by:

* Identifying unclear instructions
* Adding context and constraints
* Defining expected output format
* Clarifying intent (if needed)

👉
“My goal is to reduce interpretation gaps so the model produces consistent outputs.”

---

# 🎯 7. How do you design prompts for structured output (JSON, tables)?

**Answer:**

I explicitly define:

* Output format (JSON/table structure)
* Field names and types
* Example output (if needed)

👉 Example:
“Return the output in JSON format with fields: name, age, and city”

👉
“This ensures the output is machine-readable and consistent.”

---

# 🎯 8. How do you reduce hallucinations using prompts?

**Answer:**

To reduce hallucinations, I:

* Provide clear context
* Add constraints like “only use given information”
* Ask the model to say “I don’t know” if unsure
* Use examples to guide correct behavior

👉
“Grounding the model and limiting freedom helps reduce incorrect outputs.”

---

# 🎯 9. What is prompt sensitivity?

**Answer:**

Prompt sensitivity means small changes in wording or structure of a prompt can lead to significantly different outputs.

👉
“This is why prompt design needs consistency and careful testing.”

---

# 🎯 10. How do you control output tone and style?

**Answer:**

I control tone and style by:

* Explicitly specifying it in the prompt
* Using role-based instructions
* Adding examples if needed

👉 Example:
“Write in a formal tone for a business audience”

👉
“This ensures the output aligns with the intended audience.”

# 🧪 SECTION 2: Prompt Debugging & Optimization (CRITICAL)

# 🎯 11. A prompt is giving inconsistent results — how do you fix it?

**Answer:**

I fix inconsistency by:

* Reducing ambiguity in the prompt
* Adding clear constraints
* Defining output format
* Providing examples (few-shot)

👉
“Inconsistent outputs usually indicate that the prompt is too open-ended, so I make it more structured.”

---

# 🎯 12. How do you debug a failing prompt?

**Answer:**

My approach:

1. Identify what is failing (accuracy, format, relevance)
2. Analyze the prompt for ambiguity or missing context
3. Test with multiple inputs
4. Refine structure (role, constraints, format)
5. Re-test and iterate

👉
“I treat it like debugging — isolate the issue, fix it, and validate again.”

---

# 🎯 13. What steps do you follow to optimize a prompt?

* Improve clarity and constraints
* Add examples if needed
* Iterate based on results

👉
“Optimization is an iterative process focused on improving consistency and accuracy.”

# 🧠 SECTION 4: NLP & LLM Basics (Moderate Depth)

# 🎯 31. What is NLP?

**Answer:**

NLP (Natural Language Processing) is a field of AI that enables machines to understand, interpret, and generate human language.

👉
“It is used in applications like chatbots, text analysis, translation, and sentiment analysis.”

---

# 🎯 32. What is tokenization?

**Answer:**

Tokenization is the process of breaking text into smaller units called tokens, such as words, subwords, or characters.

👉 Example:
“Hello world” → [“Hello”, “world”]

👉
“It is the first step in processing text for NLP models.”

---

# 🎯 33. What are embeddings?

**Answer:**

Embeddings are numerical vector representations of text that capture semantic meaning.

👉
“Similar words or sentences have similar vector representations.”

👉 Use case:

* Semantic search
* Recommendation systems

---

# 🎯 34. What is Named Entity Recognition (NER)?

**Answer:**

NER is an NLP task that identifies and classifies entities in text such as names, locations, dates, or organizations.

👉 Example:
“Vatsal works at Bosch” →
Vatsal (Person), Bosch (Organization)

---

# 🎯 35. What is text classification?

**Answer:**

Text classification is the process of assigning predefined categories to text.

👉 Examples:

* Spam vs Not Spam
* Positive vs Negative sentiment

👉
“It is widely used in email filtering and sentiment analysis.”

---

# 🎯 36. What are LLMs and how do they work (high-level)?

**Answer:**

LLMs (Large Language Models) are AI models trained on large amounts of text data to understand and generate human-like language.

👉 High-level working:

* They learn patterns in language
* Predict the next word/token based on context

👉
“They don’t truly understand meaning but generate responses based on learned patterns.”

---

# 🎯 37. Why do LLMs hallucinate?

**Answer:**

LLMs hallucinate because:

* They generate responses based on probability, not facts
* Lack of sufficient context
* Over-generalization from training data

👉
“They try to give an answer even when unsure, which can lead to incorrect outputs.”

---

# 🎯 38. What is temperature in LLMs?

**Answer:**

Temperature controls randomness in model output.

* Low temperature → more predictable, consistent
* High temperature → more creative, diverse

👉
“It helps balance between accuracy and creativity.”

---

# 🎯 39. What is context window/token limit?

**Answer:**

The context window is the maximum number of tokens a model can process at once, including input and output.

👉
“If the input exceeds this limit, some information may be truncated.”

---

# 🎯 40. How do embeddings help in search or retrieval?

**Answer:**

Embeddings help by:

* Converting text into vectors
* Comparing similarity between queries and documents

👉
“This enables semantic search, where results are based on meaning rather than exact keyword matches.”
