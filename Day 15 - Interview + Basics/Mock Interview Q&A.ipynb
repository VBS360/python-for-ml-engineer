# 🤖 Technical Interview Round

## 🔹 Q1: Churn Prediction Pipeline

* Identified class imbalance (~75–25 split)
* Chose F1-score and recall over accuracy
* Applied one-hot encoding
* Built baseline model using Logistic Regression
* Handled imbalance using:

  * class_weight (performed better)
  * SMOTE (tested but less effective)
* Trained advanced models:

  * Decision Tree
  * Random Forest
* Used GridSearchCV for hyperparameter tuning
* Selected Random Forest as best model
* Derived business insights:

  * Long-term contracts reduce churn
  * Bundled services improve retention

---

## 🔹 Q2: Why class_weight performed better than SMOTE?

* class_weight adjusts learning without modifying data
* SMOTE introduces synthetic samples, which may add noise
* Logistic Regression performed better with original data distribution
* Hence, model-level balancing worked better than data-level balancing

---

## 🔹 Q3: Why Random Forest over Logistic Regression?

* Logistic Regression assumes linear relationships
* Churn depends on non-linear interactions (e.g., tenure + contract type)
* Random Forest captures feature interactions and non-linearity
* GridSearchCV further optimized performance

---

## 🔹 Q4: What does GridSearchCV do?

* Performs exhaustive search over hyperparameters
* Uses cross-validation internally
* Provides robust and reliable parameter selection
* Avoids manual trial-and-error

---

## 🔹 Q5: Why cross-validation over train-test split?

* Single split can give misleading results
* Cross-validation evaluates model across multiple splits
* Provides stable and reliable performance estimate
* Reduces overfitting risk

---

## 🔹 Q6: Overfitting scenario (95% train, 70% test)

* Indicates overfitting
* Solutions:

  * Reduce model complexity
  * Apply regularization
  * Perform feature selection
  * Use cross-validation
  * Use ensemble models

---

## 🔹 Q7: Business-focused ML (Reduce churn)

* Optimize for recall
* Capture maximum churn customers
* Apply threshold tuning / class_weight
* Convert predictions into retention strategies

---

## 🔹 Q8: What makes your projects different?

* Focus on experimentation (class_weight vs SMOTE)
* Model comparison based on business metrics
* Derive actionable insights (not just predictions)
* Strong emphasis on business impact
